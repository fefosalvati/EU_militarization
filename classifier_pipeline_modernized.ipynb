```python
# imports + environment setup, all in one place
import os
import gc
import numpy as np
import pandas as pd
import torch

from datasets import Dataset
from setfit import SetFitModel, Trainer, TrainingArguments
from transformers import set_seed

# avoid GPU memory fragmentation on long runs and silence wandb prompts
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
os.environ["WANDB_DISABLED"] = "true"

SEED = 42
set_seed(SEED)  # seeds python, numpy, torch, cuda all at once

# free leftover GPU memory from prior runs
torch.cuda.empty_cache()
gc.collect()

print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
# load and inspect
df = pd.read_csv(DATASET_PATH, on_bad_lines="skip")
print("raw shape:", df.shape)
df.head()
df.shape
```

```python
# keep only the columns we need, coerce labels to int, drop anything that is not 0 or 1
df = df[["text", "hit"]].rename(columns={"hit": "label"})
df["label"] = pd.to_numeric(df["label"], errors="coerce")
df = df.dropna(subset=["text", "label"])
df = df[df["label"].isin([0, 1])].copy()
df["label"] = df["label"].astype(int)
df["text"] = df["text"].astype(str)

print("clean shape:", df.shape)
print(df["label"].value_counts())
```

```python
# convert to HF Dataset and do a stratified split.
# class_encode_column makes "label" a ClassLabel feature, which is what
# train_test_split needs to stratify on.
dataset = Dataset.from_pandas(df, preserve_index=False).class_encode_column("label")

split = dataset.train_test_split(test_size=0.2, seed=SEED, stratify_by_column="label")
train_ds = split["train"]
test_ds = split["test"]

print(train_ds)
print(test_ds)
print("train label counts:", np.bincount(train_ds["label"]))
print("test  label counts:", np.bincount(test_ds["label"]))
```

```python
# small helpers used below.

from sklearn.metrics import f1_score, classification_report, confusion_matrix

def to_numpy(x):
    """SetFit predict/predict_proba can return a torch tensor or a list depending on version. normalize."""
    if torch.is_tensor(x):
        return x.detach().cpu().numpy()
    return np.asarray(x)

def compute_metrics(y_pred, y_true):
    """passed to the Trainer so evaluate() returns f1 in addition to accuracy."""
    return {
        "f1_macro":  f1_score(y_true, y_pred, average="macro"),
        "f1_binary": f1_score(y_true, y_pred, average="binary", pos_label=1),
    }
```

```python
# load the embedding backbone. modernbert-embed-base is a good default; you can
# swap to e.g. "sentence-transformers/all-mpnet-base-v2" or "BAAI/bge-small-en-v1.5"
# without changing anything below.
#EMBED_MODEL = "microsoft/harrier-oss-v1-0.6b"
#EMBED_MODEL = "nomic-ai/modernbert-embed-base"
EMBED_MODEL = "codefuse-ai/F2LLM-v2-80M"

model = SetFitModel.from_pretrained(
    EMBED_MODEL,
    labels=["negative", "positive"],  # human-readable, optional
)
```

```python
# SetFit creates contrastive pairs from your labeled data.
#   num_iterations = positive/negative pairs sampled per training example
#   num_epochs     = passes over those pairs (1 is the canonical SetFit recipe)
# the standard recipe is num_epochs=1, num_iterations=20. crank num_iterations up
# if you have very few labels per class.

# Importing to implement weight decay
from setfit import SetFitModel, SetFitTrainer, TrainingArguments

args = TrainingArguments(
    output_dir="setfit-out",
    num_epochs=1,             # changed from 10 to 5
    num_iterations=20,        # Reduced from 20
    batch_size=128,            # changed from 16 to 512 (temporarily adjusted to 256 because no G4 GPU is available)
    body_learning_rate=5e-7,  # adjusted from 1e-6 to 5e-7
    head_learning_rate=2e-4,  # adjusted from 3e-3 to 1e-4
    end_to_end=False,         # freeze the body during head training, then fine-tune
    eval_strategy="epoch",
    save_strategy="no",
    logging_steps=20,
    report_to="none",
    seed=SEED,
    l2_weight=0.35             # weight decay
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    metric=compute_metrics,
    column_mapping={"text": "text", "label": "label"},
)

trainer.train()
```

```python
## trainer-level metrics
metrics = trainer.evaluate()
print(metrics)
```

```python
y_true  = np.asarray(test_ds["label"])
raw_y_pred = trainer.model.predict(test_ds["text"])
# Manually define label_to_id based on how the model was initialized
label_to_id = {"negative": 0, "positive": 1}
y_pred = np.array([label_to_id[label_str] for label_str in raw_y_pred])
y_proba = to_numpy(trainer.model.predict_proba(test_ds["text"]))

print(classification_report(y_true, y_pred, digits=3))
print("confusion matrix (rows = true, cols = pred):")
print(confusion_matrix(y_true, y_pred))
```

```python
# how confident is the model on each class?
for cls in [0, 1]:
    p = y_proba[y_true == cls, 1]   # probability assigned to class 1
    print(f"class {cls}: n={len(p):4d}  mean p(1)={p.mean():.3f}  std={p.std():.3f}")
```
